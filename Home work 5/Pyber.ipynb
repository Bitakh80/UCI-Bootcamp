```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data = "data/city_data.csv"
ride_data = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_data)
ride_df = pd.read_csv(ride_data)

# Combine the data into a single dataset
merged_df = city_df.merge(ride_df, on = 'city')

# Display the data table for preview
merged_df.head()
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types
urban_city = merged_df.loc[(merged_df["type"] == "Urban")]
suburban_city = merged_df.loc[(merged_df["type"] == "Suburban")]
rural_city = merged_df.loc[(merged_df["type"] == "Rural")]


avg_fare_urban_city = urban_city.groupby(['city'])['fare'].mean()
total_rides_urban_city = urban_city.groupby(['city']).count()
total_drivers_urban_city = urban_city.groupby(['city'])['driver_count'].value_counts()


avg_fare_suburban_city = suburban_city.groupby(['city'])['fare'].mean()
total_rides_suburban_city = suburban_city.groupby(['city']).count()
total_drivers_suburban_city = suburban_city.groupby(['city'])['driver_count'].value_counts()

avg_fare_rural_city = rural_city.groupby(['city'])['fare'].mean()
total_rides_rural_city = rural_city.groupby(['city']).count()
total_drivers_rural_city = rural_city.groupby(['city'])['driver_count'].value_counts()

plt.scatter(total_rides_urban_city["ride_id"], avg_fare_urban_city,s=total_drivers_urban_city*10,
            marker ='o', facecolors ="red", edgecolors='black',alpha = 0.5, label="Urban")

plt.scatter(total_rides_suburban_city["ride_id"], avg_fare_suburban_city,s=total_drivers_suburban_city*10,
            marker ='o', facecolors ="blue", edgecolors='black',alpha = 0.5, label="Suburban")

plt.scatter(total_rides_rural_city["ride_id"], avg_fare_rural_city,s=total_drivers_rural_city*10,
             marker ='o', facecolors ="yellow", edgecolors='black',alpha = 0.55, label="Rural")


# Incorporate the other graph properties
plt.title("pyber_starter")
plt.xlabel("total number of rides")
plt.ylabel("avarage")


# Create a legend
plt.legend(loc= 'lower right')

# Save Figure
plt.savefig(f"merged_df.png")
plt.show()

```

## Total Fares by City Type

```python
total_fare = merged_df.groupby(['type'])['fare'].sum()
# Labels for the sections of our pie chart
labels = ["Rural","Suburban","Urban" ]
# The colors of each section of the pie chart
colors = ["yellow","blue","red"]
explode = (0, 0, 0.3)
plt.title("% of Total Fares By City Types")
plt.pie(total_fare, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%",shadow=True, startangle=160)
plt.axis("equal")
plt.savefig("Images/% of Total Fares By City Types.png")
plt.show()
```

```python
# Show Figure
plt.show()
```

## Total Rides by City Type

```python

total_rides = merged_df.groupby(['type'])['ride_id'].count()
labels = ["Rural","Suburban","Urban" ]
colors = ["yellow","blue","red"]
explode = (0, 0, 0.1)
plt.title("% of Total Rides By City Types")
plt.pie(total_rides, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.savefig("Images/% of Total Rides By City Types.png")
plt.show()

```

```python
# Show Figure
plt.show()
```

## Total Drivers by City Type

```python

total_drivers = merged_df.groupby(['type'])['driver_count'].sum()

labels = ["Rural","Suburban","Urban" ]

colors = ["yellow","blue","red"]
explode = (0, 0, 0.3)
plt.title("% of Total Drivers By City Types")
plt.pie(total_drivers, explode=explode, labels=labels, colors=colors,
        autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis("equal")
plt.savefig("Images/% of Total Drivers By City Types.png")
plt.show()

```
